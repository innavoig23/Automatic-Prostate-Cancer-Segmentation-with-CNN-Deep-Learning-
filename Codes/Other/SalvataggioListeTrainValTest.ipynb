**SALVATAGGIO LISTA PAZIENTI TUMORALI E PAZIENTI SANI**

```python
import os
import numpy as np
from PIL import Image
from tqdm import tqdm

def loadImage(path):
    return np.array(Image.open(path))

listaPazienti = os.listdir(os.path.join("..", "DATASET"))

listaPazientiTumorali = []
listaPazientiSani = []

for paziente in tqdm(listaPazienti):
    listaSlice = os.listdir(os.path.join("..", "DATASET", paziente, "manual"))
    pazienteTum = False
    for slice in listaSlice:
        mask = loadImage(os.path.join("..", "DATASET", paziente, "manual", slice))
        if np.sum(mask) == 0:
            continue
        else:
            pazienteTum = True
            listaPazientiTumorali.append(paziente)
            break

    if not pazienteTum:
        listaPazientiSani.append(paziente)

# Print numero pazienti sani e tumorali
print("Numero pazienti sani: ", len(listaPazientiSani))
print("Numero pazienti tumorali: ", len(listaPazientiTumorali))

# Salvo la lista dei pazienti sani e di quelli tumorali in due file
with open("pazienti_sani.txt", "w") as f:
    for paziente in listaPazientiSani:
        f.write(paziente + "\n")

with open("pazienti_tumorali.txt", "w") as f:
    for paziente in listaPazientiTumorali:
        f.write(paziente + "\n")
```

**DIVISONE LISTA TUMORALI**

```python
import os
import shutil
import random


pathDatasetCompleto = "../DATASET/"
with open('./pazienti_tumorali.txt', 'r') as file:
    pazienti = file.readlines()
ListaPazientiTumorali = [x.strip() for x in pazienti]

# creo i file txt per i tre gruppi
fileTrain = open("./pazientiTumTrain.txt", "w")
fileVal = open("./pazientiTumVal.txt", "w")
fileTest = open("./pazientiTumTest.txt", "w")

# calcolo il numero di pazienti per ogni gruppo
numeroPazienti = len(ListaPazientiTumorali)
numeroPazientiTrain = int(numeroPazienti * 0.7)
numeroPazientiVal = int(numeroPazienti * 0.15)
numeroPazientiTest = numeroPazienti - numeroPazientiTrain - numeroPazientiVal

# creo le liste di pazienti per ogni gruppo in modo random
ListaPazientiTrain = random.sample(ListaPazientiTumorali, numeroPazientiTrain)
ListaPazientiTumorali = [x for x in ListaPazientiTumorali if x not in ListaPazientiTrain]
ListaPazientiVal = random.sample(ListaPazientiTumorali, numeroPazientiVal)
ListaPazientiTumorali = [x for x in ListaPazientiTumorali if x not in ListaPazientiVal]
ListaPazientiTest = ListaPazientiTumorali

# scrivo i nomi dei pazienti nei file txt
for paziente in ListaPazientiTrain:
    fileTrain.write(paziente + "\n")

for paziente in ListaPazientiVal:
    fileVal.write(paziente + "\n")

for paziente in ListaPazientiTest:
    fileTest.write(paziente + "\n")

# chiudo i file txt
fileTrain.close()
fileVal.close()
fileTest.close()
```
